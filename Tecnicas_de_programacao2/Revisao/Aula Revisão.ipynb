#### Base de Dados

```python
# base de dados de carros
mpg = sns.load_dataset('mpg')
mpg.head(2)
```

```python
# base de dados de voos
flights = sns.load_dataset('flights')
flights.head(2)
```

#### Gráfico de Barras

* Comparativo entre dados categóricos
* Comparativos mensais e anuais
* Exemplo: Comparar o total de vendas por estado

```python
 # quantidade de carro por origem
mpg.head()
```

```python
# pandas
mpg.origin.value_counts().plot.bar()
```

```python
# matplotlib
x_bar = mpg.origin.value_counts().index
y_bar = mpg.origin.value_counts().values
plt.bar(x_bar, y_bar);
```

```python
# seaborn
sns.countplot(data=mpg, x='origin');
```

```python
# plotly
px.bar(data_frame=mpg, x='origin')
```

```python
# plotly
x_bar = mpg.origin.value_counts().index
y_bar = mpg.origin.value_counts().values
px.bar(data_frame=mpg, x=x_bar, y=y_bar)
```

```python
# plotly
px.histogram(data_frame=mpg, x='origin')
```

```python
# seaborn
# Analise bivariada (origin x aceleração)
sns.barplot(data=mpg, 
            x='origin', 
            y='acceleration', 
            estimator=np.mean, 
            errorbar=None);
```

#### Gráfico de Linhas

* Evolução de um atributo no tempo
* Demonstrar sequência de um valor durante anos, meses, semanas ou dias
* Exemplo: Vendas durante os meses de um determinado ano


```python
# evolução de voos por mês ao longo dos anos
flights.head()
```

```python
# pivotamento de flights
flights.pivot('year', 'month', 'passengers')
```

```python
# pandas
flights.pivot('year', 'month', 'passengers').plot.line();
```

```python
piv_flight = flights.pivot('year', 'month', 'passengers')
plt.plot(piv_flight)
plt.show()
```

```python
# matplotlib
piv_flight = flights.pivot('year', 'month', 'passengers')
for col in piv_flight.columns:
    plt.plot(piv_flight[col], label=col)
plt.legend()
plt.show()
```

```python
# seaborn
sns.lineplot(data=flights, 
             x='year', 
             y='passengers', 
             estimator='sum', 
             hue='month');
```

```python
# plotly
px.line(data_frame=flights,
        x='year',
        y='passengers',
        line_group='month',
        color='month')
```

#### Gráfico de Dispersão

* Visão de comparação entre duas variáveis
* Indica o quantos uma variável pode ou não influenciar a outra
* Exemplo: Tempo de profissão e salário

```python
# potência x aceleração dos carros
mpg.head()
```

```python
# pandas
mpg.plot.scatter(x='horsepower', y='acceleration');
```

```python
# matplotlib
x_scatter = mpg.horsepower
y_scatter = mpg.acceleration
plt.scatter(x_scatter, y_scatter)
plt.show()
```

```python
# seaborn
sns.scatterplot(data=mpg, x='horsepower', y='acceleration');
```

```python
# plotly
px.scatter(data_frame=mpg, 
           x='horsepower', 
           y='acceleration')
```

#### Histograma

* Visão de distribuição de uma variável
* Analisa a frequência/quantidade de uma coluna
* Similar ao gráfico de coluna, mas busca responder perguntas sobre a distribuição
* Exemplo: Faixa etária dos inscritos no enem 

```python
# verificando o aceleração dos carros
mpg.head()
```

```python
# pandas
mpg['acceleration'].plot.hist();
```

```python
# matplotlib
plt.hist(mpg.acceleration)
plt.show()
```

```python
# seaborn
sns.histplot(data=mpg, x='acceleration');
```

```python
# plotly
px.histogram(data_frame=mpg, x='acceleration')
```

#### Gráfico de Caixa

* Visão de distribuição de uma variável
* Identificação de outliers
* Apresenta algumas respostas importantes sobre o dado em análise
* Exemplo: A distribuição dos salários em uma empresa

```python
# dispersão dos dados de potência dos carros
mpg.horsepower
```

```python
# pandas
mpg.horsepower.plot.box()
```

```python
# matplolib
# no boxplot do matplolib preciso dropar os valores nulos
plt.boxplot(mpg.horsepower.dropna(), vert=True)
plt.show()
```

```python
# seaborn
sns.boxplot(data=mpg, y='horsepower');
```

```python
# plotly
px.box(data_frame=mpg, y='horsepower')
```
